```python
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
  
class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
 
        current_node = self.head
        while(current_node.next):
            current_node = current_node.next
 
        current_node.next = new_node

    def FindEvens(self):
        evens = []
        current = self.head
        while current:
            if current.data % 2 == 0:
                evens.append(current.data)
            current = current.next
        return evens



    def FindPrimes(self):
        primes = []
        current = self.head
        while current:
            if is_prime(current.data):
                primes.append(current.data)
            current = current.next
        return primes
    

    def to_list(self):
        result = []
        current = self.head
        while current:
            result.append(current.data)
            current = current.next
        return result

    def from_list(self, data_list):
        for data in data_list:
            self.append(data)

    def sort(self):
        data_list = self.to_list()
        data_list = sorted(data_list)
        self.head = None
        self.from_list(data_list)

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")
    
```

```python
llist = LinkedList()
 

llist.append(3)
llist.append(2)
llist.append(5)
llist.append(11)
llist.append(18)
llist.append(17)
llist.append(1)
llist.append(0)
llist.append(-8)

prime_numbers =FindPrimes(llist)
print("Prime numbers in the linked list:", prime_numbers)

even_numbers = llist.FindEvens()
print("\nEven numbers in the linked list:", even_numbers)

print("\nOriginal linked list:")
llist.display()

llist.sort()
print("\nSorted linked list:")
llist.display()
```
