```python
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
  
class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
 
        current_node = self.head
        while(current_node.next):
            current_node = current_node.next
 
        current_node.next = new_node

    def FindEvens(self):
        evens = []
        current = self.head
        while current:
            if current.data % 2 == 0:
                evens.append(current.data)
            current = current.next
        return evens



    def FindPrimes(self):
        primes = []
        current = self.head
        while current:
            if is_prime(current.data):
                primes.append(current.data)
            current = current.next
        return primes
    

    def to_list(self):
        result = []
        current = self.head
        while current:
            result.append(current.data)
            current = current.next
        return result

    def from_list(self, data_list):
        for data in data_list:
            self.append(data)

    def sort(self):
        data_list = self.to_list()
        data_list = sorted(data_list)
        self.head = None
        self.from_list(data_list)

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")
    
```

```python
llist = LinkedList()
 

llist.append(3)
llist.append(2)
llist.append(5)
llist.append(11)
llist.append(18)
llist.append(17)
llist.append(1)
llist.append(0)
llist.append(-8)

prime_numbers =FindPrimes(llist)
print("Prime numbers in the linked list:", prime_numbers)

even_numbers = llist.FindEvens()
print("\nEven numbers in the linked list:", even_numbers)

print("\nOriginal linked list:")
llist.display()

llist.sort()
print("\nSorted linked list:")
llist.display()
```

```python
class Book:
    def __init__(self,id,title,author,price,rating):
        self.id=id
        self.title=title
        self.author=author
        self.price=price 
        self.rating=rating
class Library:
    def __init__(self):
        self.books=[]


    def add_book(self,book):
        self.books.append(book)

    def find_by_id(self, book_id):
        for book in self.books:
            if book.id==book_id:
                print(f"{book.id}   {book.title}    {book.author}   {book.price}    {book.rating}")
        return None
    

    def find_by_author(self, author_name):
        for book in self.books:
            if book.author==author_name:
                print(f"{book.id}   {book.title}    {book.author}   {book.price}    {book.rating}")
        return None

    def find_by_rating(self, start, end):
        for book in self.books:
            if (book.rating>start and book.rating<end):
                print(f"{book.id}   {book.title}    {book.author}   {book.price}    {book.rating}")
        return None
    

    def find_by_price(self, start, end):
        for book in self.books:
            if (book.price>start and book.price<end):
                print(f"{book.id}   {book.title}    {book.author}   {book.price}    {book.rating}")
        return None

    def list_books(self):
        for book in self.books:
            self.display_book_info(book)

    def info(self):
        for i in self.books:
            print(f"{i.id}  {i.title}   {i.author}  {i.price}   {i.rating}")

    
```

```python
b1 = Book(1, "Python", "John", 250, 4.5)
b2 = Book(2, "Web", "Ana", 200, 2.0)
b3 = Book(3, "Data Science", "Alex", 320, 4.8)
b4 = Book(4, "Algorithms", "Hayley", 350, 4.2)

library = Library()
library.add_book(b1)
library.add_book(b2)
library.add_book(b3)
library.add_book(b4)

print("Library of Books:")
library.info()


print("\nFinding book by ID = 4:")
library.find_by_id(4)

print("\nFinding book by Author = Alex:")
library.find_by_author("Alex")

print("\nFinding books by rating in the range 3-5 :")
library.find_by_rating(3,5)

print("\nFinding books by price in the range = 300-500:")
library.find_by_price(300,500)

```
